### Pipelines

```python
# limpa telefone
telefone = '+55 21 3323-4432'
telefone_limpo = telefone.replace('-','').replace(' ','').replace('+','').replace('(','').replace(')','')
telefone_limpo
```

```python
# ajusta formato
numero_telefone = telefone_limpo[-8:]
ddd = telefone_limpo[-10:-8]
ddi = telefone_limpo[-14:-10]
resultado = ddi + '(' + ddd + ')' + numero_telefone[:-4] + '-' + numero_telefone[-4:]

resultado
```

```python
def limpa_telefone(telefone):
  telefone_limpo = telefone.replace('-','').replace(' ','').replace('+','').replace('(','').replace(')','')
  return telefone_limpo
```

```python
def ajusta_formato(telefone):
  numero_telefone = telefone[-8:]
  ddd = telefone[-10:-8]
  ddi = telefone[-14:-10]
  resultado = ddi + '(' + ddd + ')' + numero_telefone[:-4] + '-' + numero_telefone[-4:]
  return resultado
```

```python
num_telefone = limpa_telefone('+55 21 3323-4432')
num_telefone = ajusta_formato(num_telefone)
```

```python
num_telefone
```

```python
num_telefone = ajusta_formato(limpa_telefone('+55 21 3323-4432'))
```

```python
num_telefone
```

```python
def pipeline_telefone(telefone):
  num_telefone = limpa_telefone(telefone)
  return ajusta_formato(num_telefone)
```

```python
pipeline_telefone('+55 21 3323-4432')
```

```python
pipeline_telefone('+35 10 3323-4432')
```

### Donuts

Dado um contador inteiro do numero de donuts, retorne uma string com o formato 'Number of donuts: ' onde é o numero recebido.

Entretanto, se o parametro for 10 ou mais, use a palavra 'many' ao invés do parametro.

```python
def donuts(numero):
  if (type(numero) == int):
    if numero >= 10:
      numero = 'many'
    print(f'Number of donuts: {numero}')
  else:
    print('Precisa ser um inteiro')
```

```python
donuts(5)
```

```python
for i in range(12):
  donuts(i)
```

### both_ends

Dada uma string s, retorne uma string feita com os dois primeiros e os dois ultimos caracteres da string original.

    Exemplo: 
    'spring' retorna 'spng'. Entretanto, se o tamanho da string for menor que 2, retorne uma string vazia.

### not_bad

Dada uma string, encontre a primeira aparição das substrings 'not bad', substitua por 'good' e retorne a string resultante.

    Exemplo: 
    'The dinner is not bad!' retorna 'The dinner is good!'
