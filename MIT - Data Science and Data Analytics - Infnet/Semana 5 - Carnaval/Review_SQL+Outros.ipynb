```python
# < 8 barato
# 8 - 12 normal
# > 12 caro

df = consulta("""
  SELECT total,
    CASE
      WHEN total < 8 THEN "Barato"
      WHEN total < 12 THEN "Normal"
      WHEN total >= 12 THEN "Caro"
    END AS categoria_invoice, billing_country
   FROM invoices
  """)
```
