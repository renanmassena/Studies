```python
cat1 = [1, 1, 1]
cat2 = [1, 1, 1]
cat3 = [1, 0, 0]

dog1 = [0, 1, 0]
dog2 = [0, 0, 0]
dog3 = [0, 1, 0]
```

```python
dados = [cat1, cat2, cat3, dog1, dog2, dog3]
```

```python
dados
```

```python
classes = [1, 1, 1, 0, 0, 0]
```

```python
from sklearn.svm import LinearSVC
```

```python
model = LinearSVC()
```

```python
model.fit(dados, classes)
```

```python
animal_misterioso = [0, 0, 1]
```

```python
model.predict([animal_misterioso])
```

```python
misterioso1 = [0, 0, 1]
misterioso2 = [1, 1, 1]
misterioso3 = [1, 0, 1]
```

```python
teste_classes = [0, 1, 0]
```

```python
testes = [misterioso1, misterioso2, misterioso3]
```

```python
previsoes = model.predict(testes)
```

```python
(previsoes == teste_classes).sum()
```

```python
corretos = (previsoes == teste_classes).sum()
```

```python
corretos
```

```python
total = len(testes)
```

```python
total
```

```python
acuracia = corretos/total
```

```python
print('Taxa de Acerto: ', acuracia * 100)
```

```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')
dados = pd.read_csv('/content/drive/MyDrive/colab/tracking.csv')
```

```python
dados
```

```python
name_map = {
    'home': 'principal',
    'how_it_works': 'como_funciona',
    'contact': 'contato',
    'bought': 'comprou'
}
```

```python
dados = dados.rename(columns = name_map)
```

```python
dados
```

```python
dados.iloc[:, :-1]
```

```python
X = dados.iloc[:, :-1]
Y = dados['comprou']
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts(normalize=True) * 100
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42, stratify=Y)
```

```python
# Sem stratify
y_train.value_counts()
```

```python
y_train.value_counts()
```

```python
# Sem stratify
y_test.value_counts()
```

```python
y_test.value_counts()
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
modelo = LinearSVC()
```

```python
modelo.fit(x_train, y_train)
```

```python
yHat = modelo.predict(x_test)
```

```python
yHat
```

```python
from sklearn.metrics import accuracy_score
```

```python
acuracia = accuracy_score(yHat, y_test) * 100
```

```python
print('Aacurácia do nosso modelo foi de: ', acuracia)
```
